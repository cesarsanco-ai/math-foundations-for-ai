### Ejercicio 1.1: Límite fundamental

Calcula:
$$\lim_{x \to 0} \frac{\sin(2x)}{x}$$

```python
# Definimos la función
f1 = sp.sin(2*x) / x

# Calculamos el límite
lim1 = sp.limit(f1, x, 0)

print(f"f(x) = {f1}")
print(f"Límite cuando x→0: {lim1}")

# Verificación numérica
x_vals = np.linspace(-0.5, 0.5, 100)
y_vals = np.sin(2*x_vals) / x_vals

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, 'b-', linewidth=2)
plt.scatter(0, lim1, color='red', s=100, zorder=5, label=f'Límite = {lim1}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'$\lim_{x \to 0} \frac{\sin(2x)}{x} = 2$')
plt.legend()
plt.grid(True)
plt.axhline(0, color='gray', linestyle='--', alpha=0.5)
plt.axvline(0, color='gray', linestyle='--', alpha=0.5)
plt.show()
```

### Ejercicio 1.2: Crecimiento exponencial vs polinomial

Calcula:
$$\lim_{x \to \infty} \frac{e^x}{x^2}$$

Este límite es importante en teoría de la información y en el análisis de la función softmax.

```python
# Definimos la función
f2 = sp.exp(x) / x**2

# Calculamos el límite
lim2 = sp.limit(f2, x, sp.oo)

print(f"f(x) = {f2}")
print(f"Límite cuando x→∞: {lim2}")

# Visualización del crecimiento
x_vals = np.linspace(1, 5, 100)
y_vals = np.exp(x_vals) / x_vals**2

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, 'r-', linewidth=2)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'$\lim_{x \to \infty} \frac{e^x}{x^2} = \infty$ (crece más rápido)')
plt.grid(True)
plt.show()

print("📌 La exponencial crece más rápido que cualquier polinomio. Esto es clave en softmax: los exponenciales dominan.")
```

### Ejercicio 1.3: Límite trigonométrico importante

Calcula:
$$\lim_{x \to 0} \frac{1 - \cos x}{x^2}$$

Este límite aparece en el análisis de funciones de pérdida y en aproximaciones de segundo orden.

```python
# Definimos la función
f3 = (1 - sp.cos(x)) / x**2

# Calculamos el límite
lim3 = sp.limit(f3, x, 0)

print(f"f(x) = {f3}")
print(f"Límite cuando x→0: {lim3}")

# Verificación numérica
x_vals = np.linspace(-1, 1, 200)
y_vals = (1 - np.cos(x_vals)) / x_vals**2

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, 'g-', linewidth=2)
plt.scatter(0, lim3, color='red', s=100, zorder=5, label=f'Límite = {float(lim3):.2f}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'$\lim_{x \to 0} \frac{1 - \cos x}{x^2} = \frac{1}{2}$')
plt.legend()
plt.grid(True)
plt.show()
```

### Ejercicio 1.4: Límites laterales de una función a trozos

Define la función:
$$
g(x) = \begin{cases} 
x^2 & \text{si } x < 1 \\
2x - 1 & \text{si } x \geq 1 
\end{cases}
$$

Calcula los límites laterales en $x=1$ y determina si el límite bilateral existe.

```python
# Definimos la función a trozos en SymPy
g = sp.Piecewise((x**2, x < 1), (2*x - 1, x >= 1))

# Límites laterales
lim_izq = sp.limit(g, x, 1, dir='-')
lim_der = sp.limit(g, x, 1, dir='+')

print(f"g(x) = {g}")
print(f"Límite por izquierda (x→1⁻): {lim_izq}")
print(f"Límite por derecha (x→1⁺): {lim_der}")

if lim_izq == lim_der:
    print(f"✅ El límite bilateral existe y vale {lim_izq}")
else:
    print(f"❌ El límite bilateral NO existe (laterales diferentes)")

# Visualización
x_izq = np.linspace(0, 1, 100, endpoint=False)
x_der = np.linspace(1, 2, 100)
y_izq = x_izq**2
y_der = 2*x_der - 1

plt.figure(figsize=(8, 5))
plt.plot(x_izq, y_izq, 'b-', linewidth=2, label='x² (x<1)')
plt.plot(x_der, y_der, 'r-', linewidth=2, label='2x-1 (x≥1)')
plt.scatter(1, lim_izq, color='blue', s=100, zorder=5, label=f'Límite izq = {lim_izq}')
plt.scatter(1, lim_der, color='red', s=100, zorder=5, label=f'Límite der = {lim_der}')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.title('Función a trozos y sus límites laterales')
plt.legend()
plt.grid(True)
plt.show()
```

## Actividad 2: La Pérdida 0-1 y su Problema para Optimización

### 2.1. Definición de la pérdida 0-1

Para un problema de clasificación binaria con etiquetas $y \in \{0,1\}$ y predicción $\hat{y} \in \{0,1\}$, la **pérdida 0-1** se define como:

$$
L_{0-1}(y, \hat{y}) = \begin{cases} 
0 & \text{si } y = \hat{y} \\
1 & \text{si } y \neq \hat{y}
\end{cases}
$$

En términos de la salida continua del modelo (por ejemplo, el logit), podemos definir una función que depende del margen.

```python
# Definimos la pérdida 0-1 en función del margen (para un clasificador lineal)
# Supongamos que la predicción es signo(z) donde z es la salida lineal.
# Pérdida 0-1: 0 si y*z > 0, 1 si y*z < 0, y en z=0 es indeterminado (pero podemos definir 0.5).

def loss_01(margin):
    """Pérdida 0-1 en función del margen (y * f(x))."""
    return np.where(margin <= 0, 1, 0)

# Definimos otras pérdidas comunes para comparar
def hinge_loss(margin):
    return np.maximum(0, 1 - margin)

def logistic_loss(margin):
    return np.log(1 + np.exp(-margin))

def mse_loss(margin, y=1):
    # Para simplificar, asumimos y=1 y predicción continua
    return (1 - margin)**2

# Creamos un rango de márgenes
margin = np.linspace(-2, 2, 400)

plt.figure(figsize=(12, 6))
plt.plot(margin, loss_01(margin), 'k-', linewidth=3, label='0-1 Loss')
plt.plot(margin, hinge_loss(margin), 'b-', linewidth=2, label='Hinge Loss (SVM)')
plt.plot(margin, logistic_loss(margin), 'g-', linewidth=2, label='Logistic Loss (Cross-Entropy)')
plt.plot(margin, mse_loss(margin), 'r-', linewidth=2, label='MSE (regresión)')
plt.xlabel('Margen (y * f(x))')
plt.ylabel('Pérdida')
plt.title('Comparación de Funciones de Pérdida')
plt.legend()
plt.grid(True)
plt.axhline(0, color='gray', linestyle='--', alpha=0.5)
plt.axvline(0, color='gray', linestyle='--', alpha=0.5)
plt.xlim(-2, 2)
plt.ylim(-0.5, 4)
plt.show()
```

### 2.2. Análisis de la Pérdida 0-1

**Problemas de la pérdida 0-1 para optimización basada en gradientes:**

1.  **Discontinuidad:** La función es discontinua en el margen 0. Tiene un salto de 0 a 1.
2.  **Derivada cero en casi todos lados:** La derivada es 0 en todos los puntos donde está definida (excepto en la discontinuidad, donde no existe). Esto imposibilita el uso de gradiente descendente.
3.  **No es convexa:** Aunque en este caso es constante a trozos, no es convexa en el sentido útil para optimización.

Veamos esto con más detalle.

```python
# Analizamos la continuidad y derivabilidad
margin_fine = np.linspace(-0.5, 0.5, 1000)
loss_01_fine = loss_01(margin_fine)

plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(margin_fine, loss_01_fine, 'k-', linewidth=2)
plt.scatter(0, 0, color='blue', s=100, zorder=5, label='Valor en 0?')
plt.scatter(0, 1, color='red', s=100, zorder=5)
plt.xlabel('Margen')
plt.ylabel('Pérdida 0-1')
plt.title('Discontinuidad de la pérdida 0-1')
plt.legend(['0-1 Loss', 'Límite izquierdo?', 'Límite derecho?'])
plt.grid(True)
plt.axhline(0, color='gray', linestyle='--', alpha=0.5)
plt.axvline(0, color='gray', linestyle='--', alpha=0.5)

plt.subplot(1, 2, 2)
# Calculamos una aproximación de la derivada (diferencias finitas)
h = 0.001
deriv_approx = (loss_01(margin_fine + h) - loss_01(margin_fine - h)) / (2*h)
plt.plot(margin_fine, deriv_approx, 'r-', linewidth=2)
plt.xlabel('Margen')
plt.ylabel('Derivada aproximada')
plt.title('Derivada de la pérdida 0-1 (cero excepto en la discontinuidad)')
plt.grid(True)
plt.ylim(-1, 1)

plt.suptitle('Problemas de la pérdida 0-1 para optimización', fontsize=14)
plt.tight_layout()
plt.show()

print("📌 La derivada es 0 en casi todos lados. ¡No hay gradiente que seguir!")
```

### 2.3. Relajaciones continuas de la pérdida 0-1

Para hacer factible la optimización, se utilizan **relajaciones convexas y continuas** de la pérdida 0-1:

*   **Hinge Loss:** $\max(0, 1 - \text{margen})$ (usada en SVM).
*   **Logistic Loss:** $\log(1 + e^{-\text{margen}})$ (usada en regresión logística).
*   **Exponential Loss:** $e^{-\text{margen}}$ (usada en AdaBoost).

Todas estas son **superiores** (upper bounds) a la pérdida 0-1, lo que significa que minimizarlas también minimiza la 0-1.

```python
# Visualización de las relajaciones
margin = np.linspace(-2, 2, 400)

plt.figure(figsize=(10, 6))
plt.plot(margin, loss_01(margin), 'k--', linewidth=2, label='0-1 Loss (discontinua)')
plt.plot(margin, hinge_loss(margin), 'b-', linewidth=2, label='Hinge Loss')
plt.plot(margin, logistic_loss(margin), 'g-', linewidth=2, label='Logistic Loss')
plt.plot(margin, np.exp(-margin), 'r-', linewidth=2, label='Exponential Loss')
plt.xlabel('Margen (y * f(x))')
plt.ylabel('Pérdida')
plt.title('Relajaciones Continuas de la Pérdida 0-1')
plt.legend()
plt.grid(True)
plt.axhline(0, color='gray', linestyle='--', alpha=0.5)
plt.axvline(0, color='gray', linestyle='--', alpha=0.5)
plt.xlim(-2, 2)
plt.ylim(-0.5, 4)
plt.show()

print("✅ Todas estas funciones son continuas y diferenciables (casi en todas partes).")
```

### 2.4. Conexión con Deep Learning

En deep learning, la función de pérdida más común para clasificación es la **entropía cruzada** (binary cross-entropy), que es una versión de la logistic loss. ¿Por qué?

*   Es **diferenciable** (permite backpropagation).
*   Es **convexa** (en la salida de la red, aunque la red completa no lo sea).
*   Tiene **gradientes informativos** incluso cuando el modelo está lejos de la solución.
*   Está **bien calibrada** (produce probabilidades).

La pérdida 0-1, en cambio, no da información sobre *cuán* equivocado está el modelo, solo si acertó o no.

```python
# Ejemplo: comparación de gradientes cerca de la frontera
margin_close = np.linspace(-0.5, 0.5, 100)

# Derivadas analíticas (simplificadas)
def grad_hinge(m):
    return -1 * (m < 1)  # -1 si m<1, 0 si m>1

def grad_logistic(m):
    return -1 / (1 + np.exp(m))  # derivada de logistic loss

plt.figure(figsize=(10, 5))
plt.plot(margin_close, grad_hinge(margin_close), 'b-', linewidth=2, label='Gradiente Hinge')
plt.plot(margin_close, grad_logistic(margin_close), 'g-', linewidth=2, label='Gradiente Logistic')
plt.axhline(0, color='gray', linestyle='--', alpha=0.5)
plt.axvline(0, color='gray', linestyle='--', alpha=0.5)
plt.xlabel('Margen')
plt.ylabel('Gradiente')
plt.title('Gradientes de funciones de pérdida cerca de la frontera')
plt.legend()
plt.grid(True)
plt.show()

print("📌 La pérdida 0-1 no tiene gradiente. Las otras sí, y nos indican la dirección para mejorar.")
```
