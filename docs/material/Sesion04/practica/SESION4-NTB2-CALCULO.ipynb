```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
```

```python
# Ejercicio 1: calculo base del tema (sesion 4)
values = np.array([1, 2, 3, 4, 5], dtype=float)
resultado = values.mean() + 4 * 0.1
print('Resultado:', round(resultado, 4))
```

```python
# Ejercicio 2: simulacion mini-batch IA
y_true = np.array([1, 0, 1, 1, 0], dtype=float)
y_pred = np.array([0.9, 0.2, 0.8, 0.6, 0.3], dtype=float)
mse = np.mean((y_true - y_pred) ** 2)
print('MSE:', round(float(mse), 4))
```

```python
# Ejercicio 3: algebra lineal / tensores para modelos
A = np.array([[2., 1.], [0., 3.]])
x = np.array([1., 2.])
Ax = A @ x
print('A @ x =', Ax)
```

```python
# Ejercicio 4: visualizacion de funcion y gradiente numerico
def f(w):
    return (w - 2.0) ** 2

w = np.linspace(-1, 5, 200)
plt.plot(w, f(w))
plt.title('Funcion de costo de ejemplo')
plt.xlabel('w')
plt.ylabel('f(w)')
plt.grid(True)
plt.show()

w0 = 3.0
eps = 1e-5
grad_num = (f(w0 + eps) - f(w0 - eps)) / (2 * eps)
print('Gradiente numerico en w=3:', round(float(grad_num), 4))
```
