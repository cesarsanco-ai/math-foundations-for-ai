## 1. Derivadas Parciales: Sensibilidad Individual

Para una función de dos variables $J(w_1, w_2)$, la derivada parcial respecto a $w_1$ mide cómo cambia $J$ cuando variamos $w_1$ manteniendo $w_2$ constante.

### Ejemplo: $J(w_1, w_2) = w_1^2 + w_2^2$ (función convexa)

```python
# Definimos la función simbólicamente
J1 = w1**2 + w2**2

# Derivadas parciales
dJ1_dw1 = sp.diff(J1, w1)
dJ1_dw2 = sp.diff(J1, w2)

print(f"J(w₁, w₂) = {J1}")
print(f"∂J/∂w₁ = {dJ1_dw1}")
print(f"∂J/∂w₂ = {dJ1_dw2}")

# Evaluamos en un punto específico
punto = {w1: 1.5, w2: -1.0}
grad_en_punto = (dJ1_dw1.subs(punto), dJ1_dw2.subs(punto))
print(f"\nGradiente en (1.5, -1.0): ∇J = {grad_en_punto}")
```

### Ejemplo 2: $J(w_1, w_2) = w_1^2 - w_2^2$ (punto de silla)

```python
# Segunda función
J2 = w1**2 - w2**2

# Derivadas parciales
dJ2_dw1 = sp.diff(J2, w1)
dJ2_dw2 = sp.diff(J2, w2)

print(f"J(w₁, w₂) = {J2}")
print(f"∂J/∂w₁ = {dJ2_dw1}")
print(f"∂J/∂w₂ = {dJ2_dw2}")

# Evaluamos en el mismo punto
grad_en_punto2 = (dJ2_dw1.subs(punto), dJ2_dw2.subs(punto))
print(f"\nGradiente en (1.5, -1.0): ∇J = {grad_en_punto2}")
```

## 2. Visualización del Gradiente en 2D

### 2.1. Función convexa: $J(w_1, w_2) = w_1^2 + w_2^2$

Esta función tiene un mínimo global en (0,0). El gradiente apunta hacia afuera (dirección de máximo crecimiento), y su negativo apunta hacia el mínimo.

```python
# Definimos la función numérica y su gradiente
def J_conv(w1, w2):
    return w1**2 + w2**2

def grad_conv(w1, w2):
    return (2*w1, 2*w2)

# Creamos la malla
w1_vals = np.linspace(-3, 3, 20)
w2_vals = np.linspace(-3, 3, 20)
W1, W2 = np.meshgrid(w1_vals, w2_vals)
Z = J_conv(W1, W2)

# Gradiente en cada punto de la malla
U, V = grad_conv(W1, W2)

plt.figure(figsize=(12, 5))

# Mapa de contorno
plt.subplot(1, 2, 1)
contour = plt.contour(W1, W2, Z, levels=20, cmap='viridis')
plt.colorbar(contour, label='J(w₁,w₂)')
plt.quiver(W1, W2, U, V, color='red', alpha=0.6, scale=500)
plt.xlabel('w₁')
plt.ylabel('w₂')
plt.title('Gradiente sobre curvas de nivel: J = w₁² + w₂²')
plt.grid(True)

# Superficie 3D
ax = plt.subplot(1, 2, 2, projection='3d')
surf = ax.plot_surface(W1, W2, Z, cmap='viridis', alpha=0.7)
ax.set_xlabel('w₁')
ax.set_ylabel('w₂')
ax.set_zlabel('J')
ax.set_title('Superficie de costo')

plt.tight_layout()
plt.show()

print("📌 Observa: Las flechas (gradiente) apuntan hacia afuera, dirección de máximo crecimiento.")
print("📌 El negativo del gradiente apunta hacia el centro (mínimo).")
```

### 2.2. Punto de silla: $J(w_1, w_2) = w_1^2 - w_2^2$

Esta función tiene un punto de silla en (0,0): es un mínimo en la dirección w₁ y un máximo en la dirección w₂.

```python
def J_silla(w1, w2):
    return w1**2 - w2**2

def grad_silla(w1, w2):
    return (2*w1, -2*w2)

# Malla
W1, W2 = np.meshgrid(w1_vals, w2_vals)
Z_silla = J_silla(W1, W2)
U_silla, V_silla = grad_silla(W1, W2)

plt.figure(figsize=(12, 5))

# Mapa de contorno
plt.subplot(1, 2, 1)
contour = plt.contour(W1, W2, Z_silla, levels=20, cmap='coolwarm')
plt.colorbar(contour, label='J(w₁,w₂)')
plt.quiver(W1, W2, U_silla, V_silla, color='red', alpha=0.6, scale=500)
plt.xlabel('w₁')
plt.ylabel('w₂')
plt.title('Gradiente sobre curvas de nivel: J = w₁² - w₂² (punto silla)')
plt.grid(True)

# Superficie 3D
ax = plt.subplot(1, 2, 2, projection='3d')
surf = ax.plot_surface(W1, W2, Z_silla, cmap='coolwarm', alpha=0.7)
ax.set_xlabel('w₁')
ax.set_ylabel('w₂')
ax.set_zlabel('J')
ax.set_title('Superficie de costo (punto silla)')

plt.tight_layout()
plt.show()

print("📌 En el punto de silla (0,0), el gradiente es cero pero no es un mínimo.")
print("📌 Las flechas apuntan hacia afuera en la dirección w₁ y hacia adentro en w₂.")
```

## 3. El Gradiente como Dirección de Máximo Crecimiento

Demostración visual: en cualquier punto, el gradiente apunta en la dirección donde la función aumenta más rápidamente.

```python
# Tomamos la función convexa y evaluamos en un punto específico
punto_ejemplo = (1.5, 1.5)
grad_punto = grad_conv(*punto_ejemplo)

# Direcciones alrededor del punto
theta = np.linspace(0, 2*np.pi, 100)
r = 0.5  # radio pequeño
valores_direccion = []

for t in theta:
    w1_dir = punto_ejemplo[0] + r * np.cos(t)
    w2_dir = punto_ejemplo[1] + r * np.sin(t)
    valores_direccion.append(J_conv(w1_dir, w2_dir))

# Encontramos la dirección de máximo crecimiento
idx_max = np.argmax(valores_direccion)
dir_max = theta[idx_max]

plt.figure(figsize=(10, 8))

# Curvas de nivel
w1_vals_fino = np.linspace(-2, 3, 100)
w2_vals_fino = np.linspace(-2, 3, 100)
W1f, W2f = np.meshgrid(w1_vals_fino, w2_vals_fino)
Zf = J_conv(W1f, W2f)
plt.contour(W1f, W2f, Zf, levels=20, cmap='viridis', alpha=0.6)

# Punto de interés
plt.plot(*punto_ejemplo, 'ro', markersize=10, label='Punto (1.5, 1.5)')

# Gradiente (dirección de máximo crecimiento)
plt.arrow(*punto_ejemplo, grad_punto[0]*0.3, grad_punto[1]*0.3,
          head_width=0.1, head_length=0.1, fc='red', ec='red', label='∇J (max crecimiento)')

# Negativo del gradiente (dirección de máximo decrecimiento)
plt.arrow(*punto_ejemplo, -grad_punto[0]*0.3, -grad_punto[1]*0.3,
          head_width=0.1, head_length=0.1, fc='blue', ec='blue', label='-∇J (max decrecimiento)')

# Dirección de máximo crecimiento medida experimentalmente
plt.arrow(*punto_ejemplo, r*np.cos(dir_max), r*np.sin(dir_max),
          head_width=0.05, head_length=0.05, fc='green', ec='green', label='Dirección max crecimiento (exploración)')

plt.xlabel('w₁')
plt.ylabel('w₂')
plt.title('El gradiente apunta en la dirección de máximo crecimiento')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

## 4. Conexiones IA

### 4.1. ML/DL: Descenso de Gradiente

El gradiente de la función de coste indica la dirección de actualización de los pesos:

$$\mathbf{w}_{\text{nuevo}} = \mathbf{w}_{\text{viejo}} - \eta \nabla J(\mathbf{w}_{\text{viejo}})$$

```python
# Simulación de descenso de gradiente en la función convexa
w = np.array([2.5, 2.0])
eta = 0.1
trayectoria = [w.copy()]

for _ in range(20):
    grad = np.array(grad_conv(*w))
    w = w - eta * grad
    trayectoria.append(w.copy())

trayectoria = np.array(trayectoria)

plt.figure(figsize=(10, 8))
plt.contour(W1f, W2f, Zf, levels=20, cmap='viridis', alpha=0.6)
plt.plot(trayectoria[:,0], trayectoria[:,1], 'r.-', linewidth=2, markersize=10)
plt.plot(trayectoria[0,0], trayectoria[0,1], 'go', markersize=12, label='Inicio')
plt.plot(trayectoria[-1,0], trayectoria[-1,1], 'bo', markersize=12, label='Final')
plt.xlabel('w₁')
plt.ylabel('w₂')
plt.title('Trayectoria del descenso de gradiente')
plt.legend()
plt.grid(True)
plt.show()
```

### 4.2. CV: Detección de Bordes (Gradiente de una Imagen)

En una imagen, el gradiente resalta los bordes (zonas de cambio brusco de intensidad).

```python
from scipy import ndimage
from skimage import data, filters

# Cargamos una imagen de ejemplo (cámara)
try:
    image = data.camera()
except:
    # Si no tenemos skimage, creamos una imagen sintética
    image = np.zeros((100, 100))
    image[30:70, 30:70] = 1
    image[40:60, 40:60] = 0
    image = image + 0.2 * np.random.randn(*image.shape)
    image = np.clip(image, 0, 1)

# Gradiente en x e y (filtros Sobel)
grad_x = ndimage.sobel(image, axis=1)
grad_y = ndimage.sobel(image, axis=0)
magnitud = np.hypot(grad_x, grad_y)

plt.figure(figsize=(14, 4))
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Imagen original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(grad_x, cmap='gray')
plt.title('Gradiente en x (bordes verticales)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(magnitud, cmap='gray')
plt.title('Magnitud del gradiente (bordes)')
plt.axis('off')

plt.suptitle('Detección de bordes mediante gradiente')
plt.tight_layout()
plt.show()
```

### 4.3. NLP: Word Embeddings

En modelos de word embeddings (Word2Vec, GloVe), el gradiente se usa para acercar vectores de palabras similares y alejar los de palabras diferentes.

```python
# Simulación conceptual: vectores de palabras en 2D
palabras = {
    'rey': np.array([0.8, 0.6]),
    'reina': np.array([0.9, 0.4]),
    'hombre': np.array([0.7, -0.2]),
    'mujer': np.array([0.8, -0.3]),
    'perro': np.array([-0.5, 0.5]),
    'gato': np.array([-0.6, 0.4])
}

plt.figure(figsize=(10, 8))
for palabra, vec in palabras.items():
    plt.plot(vec[0], vec[1], 'o', markersize=10)
    plt.annotate(palabra, vec, fontsize=12, ha='center')

# Gradiente (simulado) movería palabras similares más cerca
plt.arrow(0.8, 0.6, 0.1, -0.2, head_width=0.05, fc='red', ec='red', label='Gradiente acercaría rey y reina')
plt.arrow(-0.5, 0.5, -0.1, -0.1, head_width=0.05, fc='blue', ec='blue', label='Gradiente separaría perro de hombre')

plt.xlabel('dimensión 1')
plt.ylabel('dimensión 2')
plt.title('Word embeddings: el gradiente ajusta las posiciones')
plt.legend()
plt.grid(True)
plt.axhline(0, color='gray', linestyle='--', alpha=0.3)
plt.axvline(0, color='gray', linestyle='--', alpha=0.3)
plt.xlim(-1, 1.2)
plt.ylim(-0.8, 1)
plt.show()
```

## Conclusiones de la Sesión

*   Las **derivadas parciales** miden la sensibilidad de una función multivariable respecto a cada variable individualmente.
*   El **vector gradiente** $\nabla J$ agrupa todas las derivadas parciales y apunta en la dirección de **máximo crecimiento** de la función.
*   El **negativo del gradiente** ($-\nabla J$) apunta en la dirección de máximo decrecimiento, que es la que seguimos en el descenso de gradiente para minimizar la función de costo.
*   Visualizamos el gradiente como flechas sobre mapas de contorno, observando su ortogonalidad a las curvas de nivel.
*   Diferenciamos entre funciones **convexas** (un mínimo global) y **puntos de silla** (puntos críticos que no son mínimos).
