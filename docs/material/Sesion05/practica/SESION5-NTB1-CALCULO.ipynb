## 1. Idea Intuitiva de Límite

El límite de una función $f(x)$ cuando $x$ se aproxima a un valor $a$ es el valor al que $f(x)$ se acerca, incluso si la función no está definida en $a$.

### Ejemplo clásico: $f(x) = \frac{x^2 - 1}{x - 1}$

Esta función no está definida en $x = 1$ (división por cero), pero podemos investigar su comportamiento cerca de 1.

```python
# Definimos la función simbólicamente
f_sym = (x**2 - 1) / (x - 1)
print("f(x) =", f_sym)

# Calculamos valores numéricos cerca de 1
valores_x = [0.9, 0.99, 0.999, 1.001, 1.01, 1.1]
print("\n📊 Tabla de valores cerca de x=1:")
print("x\t\tf(x)")
for xi in valores_x:
    # Evaluamos la función numéricamente (con float para evitar división por cero simbólica)
    f_num = (xi**2 - 1) / (xi - 1)
    print(f"{xi:.3f}\t\t{f_num:.6f}")

# Calculamos el límite con SymPy
limite_f = sp.limit(f_sym, x, 1)
print(f"\n🔷 Límite cuando x→1: {limite_f}")
```

---
## 2. Límites Laterales con SymPy

Los límites laterales consideran la aproximación por la izquierda ($x \to a^-$) o por la derecha ($x \to a^+$).

```python
# Función escalón (Heaviside)
escalon = sp.Piecewise((0, x < 0), (1, x >= 0))

# Límites laterales en 0
lim_izq = sp.limit(escalon, x, 0, dir='-')
lim_der = sp.limit(escalon, x, 0, dir='+')

print(f"Función escalón: {escalon}")
print(f"Límite por izquierda (x→0⁻): {lim_izq}")
print(f"Límite por derecha (x→0⁺): {lim_der}")

if lim_izq == lim_der:
    print(f"✅ El límite bilateral existe y vale {lim_izq}")
else:
    print(f"❌ El límite bilateral NO existe (laterales diferentes)")

# Función con comportamiento asimétrico: f(x) = 1/x
f_inv = 1/x
lim_inv_izq = sp.limit(f_inv, x, 0, dir='-')
lim_inv_der = sp.limit(f_inv, x, 0, dir='+')
print(f"\nFunción 1/x:")
print(f"Límite por izquierda (x→0⁻): {lim_inv_izq}")
print(f"Límite por derecha (x→0⁺): {lim_inv_der}")
```

## 3. Límites al Infinito y Asíntotas

Estudiamos el comportamiento de las funciones cuando $x$ crece o decrece indefinidamente.

```python
# Función sigmoide
sigmoide = 1 / (1 + sp.exp(-x))

# Límites al infinito
lim_sig_inf = sp.limit(sigmoide, x, sp.oo)
lim_sig_neg_inf = sp.limit(sigmoide, x, -sp.oo)

print(f"Función sigmoide: σ(x) = 1/(1+e⁻ˣ)")
print(f"Límite cuando x → +∞: {lim_sig_inf}")
print(f"Límite cuando x → -∞: {lim_sig_neg_inf}")

# Función tangente hiperbólica
tanh_sym = sp.tanh(x)
lim_tanh_inf = sp.limit(tanh_sym, x, sp.oo)
lim_tanh_neg_inf = sp.limit(tanh_sym, x, -sp.oo)

print(f"\nFunción tanh:")
print(f"Límite cuando x → +∞: {lim_tanh_inf}")
print(f"Límite cuando x → -∞: {lim_tanh_neg_inf}")
```

```python
# Visualización de límites al infinito
x_vals = np.linspace(-10, 10, 400)
sig_vals = 1 / (1 + np.exp(-x_vals))
tanh_vals = np.tanh(x_vals)

plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(x_vals, sig_vals, 'b-', linewidth=2)
plt.axhline(y=1, color='r', linestyle='--', label='Asíntota y=1')
plt.axhline(y=0, color='g', linestyle='--', label='Asíntota y=0')
plt.xlabel('x')
plt.ylabel('σ(x)')
plt.title('Sigmoide: Límites al infinito')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x_vals, tanh_vals, 'r-', linewidth=2)
plt.axhline(y=1, color='blue', linestyle='--', label='Asíntota y=1')
plt.axhline(y=-1, color='green', linestyle='--', label='Asíntota y=-1')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.title('Tangente Hiperbólica: Límites al infinito')
plt.legend()
plt.grid(True)

plt.suptitle('Comportamiento Asintótico de Funciones de Activación', fontsize=14)
plt.tight_layout()
plt.show()
```

## 4. Visualización de Discontinuidades

### 4.1. Discontinuidad de Salto: Función Escalón

```python
# Función escalón (Heaviside) numérica
def heaviside(x):
    return np.where(x >= 0, 1, 0)

x_vals = np.linspace(-2, 2, 1000)
y_vals = heaviside(x_vals)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='H(x)')
plt.scatter(0, 0, color='red', s=100, zorder=5, label='Punto de discontinuidad')
plt.scatter(0, 1, color='red', s=100, zorder=5)
plt.xlabel('x')
plt.ylabel('H(x)')
plt.title('Discontinuidad de Salto: Función Escalón')
plt.legend()
plt.grid(True)
plt.axvline(0, color='gray', linestyle='--', alpha=0.5)
plt.show()
```

### 4.2. Discontinuidad de Salto: Función Parte Entera (piso)

```python
x_vals = np.linspace(-3, 3, 1000)
y_vals = np.floor(x_vals)  # Función piso (parte entera)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='⌊x⌋')
plt.xlabel('x')
plt.ylabel('floor(x)')
plt.title('Discontinuidad de Salto: Función Parte Entera')
plt.grid(True)
plt.ylim(-4, 4)
plt.show()
```

### 4.3. Discontinuidad Evitable (Removible)

```python
# Función con discontinuidad evitable: (x^2 - 1)/(x - 1)
def f_evitable(x):
    return (x**2 - 1) / (x - 1)

x_vals = np.linspace(0, 2, 400)
y_vals = f_evitable(x_vals)

# La función tiene un hueco en x=1
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='f(x) = (x²-1)/(x-1)')
plt.scatter(1, 2, color='red', s=100, zorder=5, label='Hueco (discontinuidad)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Discontinuidad Evitable')
plt.legend()
plt.grid(True)
plt.show()
```

### 4.4. Discontinuidad Infinita: $f(x) = 1/x$

```python
x_vals1 = np.linspace(-2, -0.1, 400)
x_vals2 = np.linspace(0.1, 2, 400)
y_vals1 = 1/x_vals1
y_vals2 = 1/x_vals2

plt.figure(figsize=(8, 5))
plt.plot(x_vals1, y_vals1, 'b-', linewidth=2, label='1/x (x<0)')
plt.plot(x_vals2, y_vals2, 'r-', linewidth=2, label='1/x (x>0)')
plt.axvline(0, color='black', linestyle='--', label='Asíntota x=0')
plt.xlabel('x')
plt.ylabel('1/x')
plt.title('Discontinuidad Infinita en x=0')
plt.legend()
plt.grid(True)
plt.ylim(-10, 10)
plt.show()
```

## 5. Comparación con Funciones Continuas a Trozos: ReLU

La función **ReLU** (Rectified Linear Unit) es continua pero no diferenciable en $x=0$. Es la función de activación más usada en deep learning.

```python
def relu(x):
    return np.maximum(0, x)

x_vals = np.linspace(-3, 3, 400)
y_vals = relu(x_vals)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='ReLU(x) = max(0, x)')
plt.scatter(0, 0, color='red', s=100, zorder=5, label='Punto no diferenciable')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.title('ReLU: Continua pero no diferenciable en x=0')
plt.legend()
plt.grid(True)
plt.axhline(0, color='gray', linestyle='--', alpha=0.5)
plt.axvline(0, color='gray', linestyle='--', alpha=0.5)
plt.show()

# Verificamos límites laterales
lim_izq_relu = sp.limit(sp.Max(0, x), x, 0, dir='-')
lim_der_relu = sp.limit(sp.Max(0, x), x, 0, dir='+')
print(f"Límite izquierdo ReLU (x→0⁻): {lim_izq_relu}")
print(f"Límite derecho ReLU (x→0⁺): {lim_der_relu}")
print(f"Valor en x=0: {relu(0)}")
print("✅ La función es continua porque límites laterales = valor en el punto.")
print("❌ No es diferenciable en 0 porque la derivada izquierda (0) ≠ derecha (1).")
```

## 6. Conexiones IA: ¿Por qué necesitamos funciones (casi) diferenciables?

### Backpropagation
El algoritmo de backpropagation se basa en la **regla de la cadena** para calcular gradientes. Para ello, las funciones de activación deben ser diferenciables (o al menos tener subgradientes).

*   **Sigmoide, tanh**: Diferenciables en todo $\mathbb{R}$. Suaves.
*   **ReLU**: Diferenciable excepto en un punto (0). En la práctica, se usa un **subgradiente** (por ejemplo, 0 para x<0, 1 para x>0, y se puede elegir 0 o 1 en x=0).
*   **Escalón**: No es diferenciable (discontinua). No se puede usar directamente con gradiente descendente.

### Convergencia de algoritmos
El análisis de convergencia del gradiente descendente asume que la función de pérdida es diferenciable (o al menos Lipschitz). Los límites nos ayudan a entender el comportamiento asintótico.

```python
# Ejemplo: Entropía cruzada y su comportamiento asintótico
p = np.linspace(0.001, 1, 100)
loss = -np.log(p)  # para y=1

plt.figure(figsize=(8, 5))
plt.plot(p, loss, 'b-', linewidth=2)
plt.xlabel('Probabilidad predicha $\hat{y}$')
plt.ylabel('Pérdida')
plt.title('Entropía Cruzada (y=1): Comportamiento asintótico')
plt.axvline(1, color='r', linestyle='--', label='$\hat{y}=1$')
plt.text(0.5, 3, 'La pérdida tiende a 0 cuando $\hat{y} \to 1$', fontsize=10)
plt.legend()
plt.grid(True)
plt.show()

# Límite simbólico
loss_sym = -sp.log(x)
lim_loss = sp.limit(loss_sym, x, 1)
print(f"Límite de -log(ŷ) cuando ŷ→1: {lim_loss}")
```
