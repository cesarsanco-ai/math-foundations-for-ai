## Configuración Inicial

Importamos las librerías necesarias y generamos datos sintéticos para experimentar.

```python
# Importamos librerías
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de gráficos
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 5)

# Fijamos semilla para reproducibilidad
np.random.seed(42)

# Generamos datos sintéticos: vectores y matrices
v1 = np.random.randint(1, 10, size=5)
v2 = np.random.randint(1, 10, size=5)
M1 = np.random.randint(1, 10, size=(3, 4))
M2 = np.random.randint(1, 10, size=(3, 4))

print("🎯 Datos sintéticos generados:")
print(f"v1 = {v1}")
print(f"v2 = {v2}")
print(f"M1 = \n{M1}")
print(f"M2 = \n{M2}")
```

## 1. Operaciones Elemento a Elemento (Element-wise)

Las operaciones elemento a elemento se aplican de forma independiente a cada par de elementos correspondientes de dos tensores del mismo shape. Son la base de la computación paralela en GPUs.

### 1.1. Suma y Resta

```python
print("🔷 SUMA ELEMENTO A ELEMENTO")
print(f"v1 = {v1}")
print(f"v2 = {v2}")
print(f"v1 + v2 = {v1 + v2}")

print("\nM1 + M2:")
print(M1 + M2)

print("\n🔶 RESTA ELEMENTO A ELEMENTO")
print(f"v1 - v2 = {v1 - v2}")
print("\nM1 - M2:")
print(M1 - M2)
```

### 1.2. Producto Hadamard (⊙)

El producto Hadamard es la multiplicación elemento a elemento. En NumPy se usa el operador `*`. Es diferente de la multiplicación de matrices.

```python
print("🔷 PRODUCTO HADAMARD (elemento a elemento)")
print(f"v1 * v2 = {v1 * v2}")

print("\nM1 * M2 (Hadamard):")
print(M1 * M2)

# Visualización del producto Hadamard
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

sns.heatmap(M1, annot=True, fmt='d', cmap='Blues', ax=axes[0], cbar=False)
axes[0].set_title('Matriz A')

sns.heatmap(M2, annot=True, fmt='d', cmap='Greens', ax=axes[1], cbar=False)
axes[1].set_title('Matriz B')

sns.heatmap(M1 * M2, annot=True, fmt='d', cmap='Purples', ax=axes[2], cbar=False)
axes[2].set_title('A ⊙ B (Hadamard)')

plt.suptitle('Producto Hadamard: Multiplicación Elemento a Elemento', fontsize=14)
plt.tight_layout()
plt.show()
```

## 2. Broadcasting

El **broadcasting** es una de las características más poderosas de NumPy. Permite realizar operaciones entre tensores de diferentes formas, "estirando" el tensor más pequeño sin copiar datos explícitamente (ahorra memoria y tiempo).

### Reglas de Broadcasting:
1. Se comparan las dimensiones desde la última hacia la primera.
2. Dos dimensiones son compatibles si:
   * Son iguales, o
   * Una de ellas es 1, o
   * Una de ellas no existe (se trata como 1).

### 2.1. Escalar + Vector

```python
escalar = 10
vector = np.array([1, 2, 3, 4])

print(f"Escalar: {escalar}, shape: {np.array(escalar).shape}")
print(f"Vector: {vector}, shape: {vector.shape}")
print(f"Resultado (escalar + vector): {escalar + vector}")
print("El escalar se 'stira' para coincidir con cada elemento del vector.")
```

### 2.2. Vector + Matriz (vector fila sobre filas)

```python
v = np.array([1, 2, 3])  # shape (3,)
M = np.array([[10, 20, 30],
              [40, 50, 60]])  # shape (2, 3)

print(f"Vector v: {v}, shape: {v.shape}")
print(f"Matriz M:\n{M}, shape: {M.shape}")
print(f"\nM + v (broadcasting sobre filas):\n{M + v}")

# Visualización del broadcasting
fig, axes = plt.subplots(1, 3, figsize=(15, 3))

sns.heatmap(M, annot=True, fmt='d', cmap='Blues', ax=axes[0], cbar=False)
axes[0].set_title('Matriz M (2x3)')

v_broadcast = np.tile(v, (2, 1))  # Mostramos explícitamente el estiramiento
sns.heatmap(v_broadcast, annot=True, fmt='d', cmap='Greens', ax=axes[1], cbar=False)
axes[1].set_title('Vector v "estirado" a (2,3)')

sns.heatmap(M + v, annot=True, fmt='d', cmap='Purples', ax=axes[2], cbar=False)
axes[2].set_title('Resultado M + v')

plt.suptitle('Broadcasting: Vector sobre filas de una Matriz', fontsize=14)
plt.tight_layout()
plt.show()
```

### 2.3. Vector columna + Matriz

```python
v_col = np.array([[1], [2]])  # shape (2, 1)
M = np.array([[10, 20],
              [30, 40]])  # shape (2, 2)

print(f"Vector columna v:\n{v_col}, shape: {v_col.shape}")
print(f"Matriz M:\n{M}, shape: {M.shape}")
print(f"\nM + v_col (broadcasting sobre columnas):\n{M + v_col}")
```

### 2.4. Aplicación en IA: Batch Normalization (simplificada)

En Batch Normalization, se normaliza cada canal de un lote de imágenes. Se resta la media del canal y se divide por la desviación estándar. Estas operaciones usan broadcasting.

```python
# Simulamos un lote de 3 imágenes RGB de 4x4
batch = np.random.randint(0, 256, size=(3, 4, 4, 3)).astype(np.float32)
print(f"Shape del lote (batch, alto, ancho, canales): {batch.shape}")

# Calculamos la media por canal a lo largo de las dimensiones espaciales (alto, ancho)
# manteniendo las dimensiones para broadcasting
media_por_canal = batch.mean(axis=(1, 2), keepdims=True)  # shape: (3, 1, 1, 3)
print(f"Shape de la media por imagen y canal: {media_por_canal.shape}")

# Restamos la media (broadcasting: (3,4,4,3) - (3,1,1,3))
batch_centrado = batch - media_por_canal
print(f"Shape después de restar media: {batch_centrado.shape}")
print("✅ Broadcasting permite restar un tensor 4D de uno 4D con dimensiones intermedias 1.")
```

## 3. Reducción: Suma y Media a lo largo de ejes

La reducción consiste en aplicar una operación (suma, media, máximo, etc.) a lo largo de uno o varios ejes, eliminando esas dimensiones.

### 3.1. Suma total y suma por ejes

```python
M = np.array([[1, 2, 3],
              [4, 5, 6]])
print(f"Matriz M:\n{M}")

print(f"\n🔷 Suma total (todos los elementos): {M.sum()}")
print(f"🔶 Suma a lo largo de filas (axis=0) - por columna: {M.sum(axis=0)}")
print(f"🔷 Suma a lo largo de columnas (axis=1) - por fila: {M.sum(axis=1)}")

print(f"\n📊 Media por columna (axis=0): {M.mean(axis=0)}")
print(f"📊 Media por fila (axis=1): {M.mean(axis=1)}")

# Visualización de la reducción
fig, axes = plt.subplots(1, 3, figsize=(15, 3))

sns.heatmap(M, annot=True, fmt='d', cmap='Blues', ax=axes[0], cbar=False)
axes[0].set_title('Matriz Original')

suma_col = M.sum(axis=0).reshape(1, -1)
sns.heatmap(suma_col, annot=True, fmt='d', cmap='Reds', ax=axes[1], cbar=False)
axes[1].set_title('Suma por columna (axis=0)')

suma_fila = M.sum(axis=1).reshape(-1, 1)
sns.heatmap(suma_fila, annot=True, fmt='d', cmap='Greens', ax=axes[2], cbar=False)
axes[2].set_title('Suma por fila (axis=1)')

plt.suptitle('Reducción: Suma a lo largo de ejes', fontsize=14)
plt.tight_layout()
plt.show()
```

### 3.2. Aplicación en IA: Pérdida promedio en un batch

Durante el entrenamiento, se calcula la pérdida para cada muestra y luego se promedia sobre el batch.

```python
# Simulamos pérdidas individuales para un batch de 32 muestras
batch_size = 32
perdidas_individuales = np.random.randn(batch_size)**2  # Pérdidas positivas

perdida_promedio = perdidas_individuales.mean()
print(f"Pérdidas individuales (primeras 5): {perdidas_individuales[:5]}")
print(f"📉 Pérdida promedio del batch: {perdida_promedio:.4f}")

# Equivalente con suma
perdida_promedio_manual = perdidas_individuales.sum() / batch_size
print(f"Pérdida promedio (manual): {perdida_promedio_manual:.4f}")
```

## 4. Producto Punto y Multiplicación de Matrices

El **producto punto** (o producto escalar) es una operación fundamental entre dos vectores que produce un escalar. Es la base de la similitud, de las proyecciones y de la atención.

### 4.1. Producto punto entre vectores

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

punto = np.dot(a, b)
punto_manual = a[0]*b[0] + a[1]*b[1] + a[2]*b[2]

print(f"a = {a}")
print(f"b = {b}")
print(f"Producto punto a·b = {punto}")
print(f"Verificación manual: {punto_manual}")
```

### 4.2. Producto matriz-vector

Una capa fully-connected en una red neuronal es exactamente esto: $y = Wx + b$.

```python
W = np.array([[1, 2, 3],   # pesos de 2 neuronas (2x3)
              [4, 5, 6]])
x = np.array([2, 3, 4])    # entrada (3,)

y = np.dot(W, x)  # o W @ x
print(f"Matriz de pesos W (2 neuronas, 3 entradas):\n{W}")
print(f"Vector de entrada x: {x}")
print(f"Salida de la capa fully-connected y = Wx: {y}")

# Con sesgo (broadcasting!)
b = np.array([0.5, -0.5])
y_con_sesgo = np.dot(W, x) + b
print(f"Con sesgo b={b}, salida: {y_con_sesgo}")
```

### 4.3. Multiplicación de matrices

El producto de matrices es la base de la propagación en lotes (batch processing).

```python
A = np.array([[1, 2],
              [3, 4],
              [5, 6]])  # 3x2
B = np.array([[7, 8, 9],
              [10, 11, 12]])  # 2x3

C = np.dot(A, B)  # o A @ B
print(f"Matriz A (3x2):\n{A}")
print(f"Matriz B (2x3):\n{B}")
print(f"Producto A·B (3x3):\n{C}")

# Visualización
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

sns.heatmap(A, annot=True, fmt='d', cmap='Blues', ax=axes[0], cbar=False)
axes[0].set_title('Matriz A (3x2)')

sns.heatmap(B, annot=True, fmt='d', cmap='Greens', ax=axes[1], cbar=False)
axes[1].set_title('Matriz B (2x3)')

sns.heatmap(C, annot=True, fmt='d', cmap='Purples', ax=axes[2], cbar=False)
axes[2].set_title('Resultado A·B (3x3)')

plt.suptitle('Multiplicación de Matrices', fontsize=14)
plt.tight_layout()
plt.show()
```

## 5. Similitud de Coseno

La **similitud de coseno** mide el ángulo entre dos vectores, independientemente de su magnitud. Es muy usada en NLP para comparar embeddings de palabras u oraciones.

$$\text{cosim}(\mathbf{a}, \mathbf{b}) = \frac{\mathbf{a} \cdot \mathbf{b}}{\|\mathbf{a}\| \|\mathbf{b}\|}$$

```python
def cos_sim(a, b):
    dot = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    return dot / (norm_a * norm_b)

# Ejemplo 1: Vectores similares (ángulo pequeño)
a1 = np.array([2, 1])
b1 = np.array([4, 2])  # b1 = 2*a1

# Ejemplo 2: Vectores ortogonales (sin similitud)
a2 = np.array([1, 0])
b2 = np.array([0, 1])

# Ejemplo 3: Vectores opuestos
a3 = np.array([1, 0])
b3 = np.array([-1, 0])

print(f"Similitud entre {a1} y {b1}: {cos_sim(a1, b1):.3f} (casi 1, misma dirección)")
print(f"Similitud entre {a2} y {b2}: {cos_sim(a2, b2):.3f} (0, ortogonales)")
print(f"Similitud entre {a3} y {b3}: {cos_sim(a3, b3):.3f} (-1, direcciones opuestas)")

# Visualización de vectores
origen = np.array([0, 0])
plt.figure(figsize=(8, 8))
plt.quiver(*origen, a1[0], a1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='a1')
plt.quiver(*origen, b1[0], b1[1], angles='xy', scale_units='xy', scale=1, color='lightblue', label='b1')
plt.quiver(*origen, a2[0], a2[1], angles='xy', scale_units='xy', scale=1, color='red', label='a2')
plt.quiver(*origen, b2[0], b2[1], angles='xy', scale_units='xy', scale=1, color='orange', label='b2')
plt.quiver(*origen, a3[0], a3[1], angles='xy', scale_units='xy', scale=1, color='green', label='a3')
plt.quiver(*origen, b3[0], b3[1], angles='xy', scale_units='xy', scale=1, color='lightgreen', label='b3')
plt.xlim(-2, 5)
plt.ylim(-2, 3)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Vectores y su similitud de coseno')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

### 6.4. GenAI: Mecanismo de Atención en Transformers
El corazón de los transformers: $\text{Attention}(Q, K, V) = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)V$

```python
# Simulación simplificada del mecanismo de atención
def simple_attention(Q, K, V):
    d_k = Q.shape[-1]
    scores = np.dot(Q, K.T) / np.sqrt(d_k)  # Productos punto
    weights = np.exp(scores) / np.sum(np.exp(scores), axis=-1, keepdims=True)  # Softmax
    output = np.dot(weights, V)
    return output, weights

# Datos sintéticos: 3 consultas (queries), 4 claves (keys), dimensión 64
Q = np.random.randn(3, 64)
K = np.random.randn(4, 64)
V = np.random.randn(4, 128)  # 4 valores, cada uno de 128 dimensiones

output, att_weights = simple_attention(Q, K, V)
print(f"Matriz de atención (pesos): shape {att_weights.shape}")
print(f"Primera fila (atención de query 0 sobre las 4 keys): {att_weights[0]:.3f}")
print(f"Suma de pesos para query 0: {att_weights[0].sum():.2f} (debe ser 1)")

# Visualización de la matriz de atención
plt.figure(figsize=(6, 5))
sns.heatmap(att_weights, annot=True, fmt='.2f', cmap='viridis')
plt.title('Matriz de Atención (Q·K^T + softmax)')
plt.xlabel('Keys')
plt.ylabel('Queries')
plt.show()
```
